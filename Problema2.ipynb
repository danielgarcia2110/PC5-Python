```python
# Renombrar columnas
df.rename(columns={
    'Unnamed: 0': 'id',
    'taster_name': 'critic',
    'taster_twitter_handle': 'critic_twitter',
    'region_1': 'region',
    'region_2': 'subregion'
}, inplace=True)

# Mapeo país -> continente
country_to_continent = {
    'US': 'North America', 'United States': 'North America', 'USA': 'North America', 'Canada': 'North America', 'Mexico': 'North America',
    'Argentina': 'South America', 'Chile': 'South America', 'Brazil': 'South America', 'Uruguay': 'South America',
    'France': 'Europe', 'Italy': 'Europe', 'Spain': 'Europe', 'Portugal': 'Europe', 'Germany': 'Europe', 'Austria': 'Europe',
    'United Kingdom': 'Europe', 'England': 'Europe', 'Switzerland': 'Europe', 'Greece': 'Europe', 'Hungary': 'Europe',
    'Romania': 'Europe', 'Bulgaria': 'Europe', 'Czech Republic': 'Europe', 'Slovenia': 'Europe',
    'Australia': 'Oceania', 'New Zealand': 'Oceania',
    'South Africa': 'Africa', 'Morocco': 'Africa',
    'Israel': 'Asia', 'Lebanon': 'Asia', 'Japan': 'Asia', 'China': 'Asia'
}

# Crear nuevas columnas
# 1) continente a partir de country
df['continent'] = df['country'].map(lambda c: country_to_continent.get(c, 'Other') if pd.notna(c) else 'Unknown')

# 2) categorización de precio
def price_category(p):
    if pd.isna(p):
        return 'Unknown'
    p = float(p)
    if p < 10:
        return 'Cheap'
    if p < 20:
        return 'Budget'
    if p < 50:
        return 'Mid'
    if p < 100:
        return 'Premium'
    return 'Luxury'
df['price_category'] = df['price'].apply(price_category)

# 3) longitud de la descripción en número de palabras
df['description_word_count'] = df['description'].str.split().str.len()

# indicador de alta puntuación
df['is_high_score'] = df['points'] >= 90

# Mostrar un resumen de las columnas nuevas y renombradas
df[['id', 'country', 'continent', 'price', 'price_category', 'points', 'is_high_score', 'description_word_count']].head()
```

```python
# 1) Top 3 vinos por continente (por puntos, en caso de empate ordenar por precio)
top3_por_continente = (
    df.sort_values(['continent', 'points', 'price'], ascending=[True, False, False])
      .groupby('continent', as_index=False)
      .head(3)
      .loc[:, ['continent', 'title', 'winery', 'country', 'points', 'price']]
)
print("Reporte 1: Top 3 vinos por continente (por points, luego price)\n")
print(top3_por_continente.to_string(index=False))
print("\n" + "="*80 + "\n")
```

```python
# 2) Promedio de precio y cantidad de reviews por país (países con >=100 reviews), ordenado por precio promedio descendente
country_stats = (
    df[df['country'].notna()]
      .groupby('country')
      .agg(reviews=('id', 'count'),
           avg_price=('price', 'mean'),
           avg_points=('points', 'mean'))
      .reset_index()
)
country_stats_filtered = (
    country_stats[country_stats['reviews'] >= 100]
      .assign(avg_price=lambda d: d['avg_price'].round(2),
              avg_points=lambda d: d['avg_points'].round(2))
      .sort_values('avg_price', ascending=False)
)
print("Reporte 2: Promedio de precio y cantidad de reviews por país (>=100 reviews), ordenado por avg_price desc\n")
print(country_stats_filtered.to_string(index=False))
print("\n" + "="*80 + "\n")
```

```python
# 3) Distribución de categorías de precio por continente (conteos y porcentaje)
pc_counts = (
    df.groupby(['continent', 'price_category'])
      .size()
      .reset_index(name='count')
)
totales_cont = df.groupby('continent').size().reset_index(name='total')
pc = pc_counts.merge(totales_cont, on='continent')
pc['percent'] = (pc['count'] / pc['total'] * 100).round(2)
pc_pivot_counts = pc.pivot(index='continent', columns='price_category', values='count').fillna(0).astype(int)
pc_pivot_percent = pc.pivot(index='continent', columns='price_category', values='percent').fillna(0).round(2)
print("Reporte 3: Conteo de price_category por continente\n")
print(pc_pivot_counts.to_string())
print("\nPorcentajes (%) por continente:\n")
print(pc_pivot_percent.to_string())
print("\n" + "="*80 + "\n")
```

```python
# 4) Top 10 variedades más reseñadas: cantidad de reviews, promedio de puntos y promedio de precio
var_stats = (
    df.groupby('variety')
      .agg(reviews=('id', 'count'),
           avg_points=('points', 'mean'),
           avg_price=('price', 'mean'))
      .reset_index()
      .assign(avg_points=lambda d: d['avg_points'].round(2),
              avg_price=lambda d: d['avg_price'].round(2))
)
top10_varieties = var_stats.sort_values('reviews', ascending=False).head(10)
print("Reporte 4: Top 10 variedades por número de reseñas (reviews), con avg_points y avg_price\n")
print(top10_varieties.to_string(index=False))
```

```python
from pathlib import Path
import os
import sqlite3

out_dir = Path("reports")
out_dir.mkdir(exist_ok=True)
```

```python
# 1) CSV: un archivo por reporte
top3_por_continente.to_csv(out_dir / "top3_por_continente.csv", index=False)
country_stats_filtered.to_csv(out_dir / "country_stats_filtered.csv", index=False)
pc_pivot_counts.to_csv(out_dir / "pc_pivot_counts.csv", index=True)  # índice = continentes
top10_varieties.to_csv(out_dir / "top10_varieties.csv", index=False)
```

```python
# 2) Excel: un workbook con una hoja por reporte
try:
    with pd.ExcelWriter(out_dir / "wine_reports.xlsx", engine="openpyxl") as writer:
        top3_por_continente.to_excel(writer, sheet_name="top3_por_continente", index=False)
        country_stats_filtered.to_excel(writer, sheet_name="country_stats_filtered", index=False)
        pc_pivot_counts.to_excel(writer, sheet_name="pc_pivot_counts")  # conserva índice (continent)
        top10_varieties.to_excel(writer, sheet_name="top10_varieties", index=False)
except Exception as e:
    print("No se pudo crear Excel (falta engine). Error:", e)
```
