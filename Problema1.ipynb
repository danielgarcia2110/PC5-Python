```python
candidates = df_airbnb[
    (df_airbnb['accommodates'] >= 4) &
    (df_airbnb['bedrooms'] >= 2) &
    (df_airbnb['reviews'] > 10) &
    (df_airbnb['overall_satisfaction'] > 4)
].copy()

# Ordenar por puntuación desc y luego por número de críticas desc
candidates = candidates.sort_values(by=['overall_satisfaction', 'reviews'], ascending=[False, False])

# Seleccionar las 3 mejores alternativas y mostrar columnas relevantes
top3 = candidates.head(3)[[
    'room_id', 'host_id', 'room_type', 'neighborhood',
    'reviews', 'overall_satisfaction', 'accommodates', 'bedrooms', 'price'
]].reset_index(drop=True)

top3
```

```python
nights = 3
budget_per_night = 50.0

eligible = df_airbnb[df_airbnb['price'] <= budget_per_night].copy()

if eligible.empty:
    total_budget = budget_per_night  # aquí budget_per_night se interpreta como presupuesto total
    eligible = df_airbnb[df_airbnb['price'] * nights <= total_budget].copy()
    mode = "presupuesto_total"
else:
    mode = "presupuesto_por_noche"

shared = eligible[eligible['room_type'] == 'Shared room'].sort_values(
    by=['overall_satisfaction', 'price'], ascending=[False, True]
)
others = eligible[eligible['room_type'] != 'Shared room'].sort_values(by='price', ascending=True)

result = pd.concat([shared, others]).head(10).copy()

result['total_cost'] = result['price'] * nights
cols = ['room_id', 'host_id', 'room_type', 'neighborhood', 'price', 'total_cost', 'reviews', 'overall_satisfaction', 'accommodates', 'bedrooms']
result = result[cols].reset_index(drop=True)

print(f"Modo de filtrado: {mode} (noches={nights}, presupuesto_por_noche={budget_per_night})")
result
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1) precio vs satisfacción
sns.scatterplot(
    data=candidates,
    x='price',
    y='overall_satisfaction',
    hue='room_type',
    size='reviews',
    sizes=(20, 200),
    alpha=0.7,
    ax=axes[0]
)
axes[0].set_title('Candidates: Precio vs Satisfacción')
axes[0].set_xlabel('Precio (por noche)')
axes[0].set_ylabel('Overall satisfaction')
axes[0].legend(loc='best', fontsize='small')

# 2) Conteo por tipo de habitación
order = result['room_type'].value_counts().index
sns.countplot(data=result, x='room_type', order=order, palette='pastel', ax=axes[1])
axes[1].set_title('Distribución por tipo (result)')
axes[1].set_xlabel('Tipo de habitación')
axes[1].set_ylabel('Cantidad')

# 3) coste total
sns.boxplot(data=result, x='room_type', y='total_cost', palette='muted', ax=axes[2])
axes[2].set_title('Total cost ({} noches) por tipo'.format(nights))
axes[2].set_xlabel('Tipo de habitación')
axes[2].set_ylabel('Coste total')

plt.tight_layout()
plt.show()
```
